### Step 1: Extracting slide content

````python
def extract_flags(text: str, startflag : str = '<', endflag : str = '>' ) -> list:
    flags = []
    start = 0
    while True:
        start = text.find(startflag, start)
        if start == -1:
            break
        end = text.find(endflag, start)
        if end == -1:
            break

        flags.append(text[start:end+1])
        start = end + 1
    
    if len(flags) % 2 != 0:
        raise Exception(f"Uneven number of separation flags. Start and End flags come in pairs")

    return flags
````

````python
flags = extract_flags(pptx_plan)
slide_content = {}
slidenum = 1
for i in range(0,len(flags),2):
    slide_content[f"slide_{slidenum}"] = find_text_in_between_tags(pptx_plan, flags[i], flags[i + 1])
    slidenum += 1
````

````python
def reviewer(review_prompt_template : str, html_txt : str, html_image : Image.Image, llm : LLM):
    
    #Sanitizing input html
    html_txt = find_text_in_between_tags(html_txt, start_tag="<!DOCTYPE html>", end_tag="</html>", inclusive=True)

    review_prompt = review_prompt_template.format(code=html_txt)
    response = llm.call_with_images(query=review_prompt, images=[html_image])

    if '<OK>' in response['text']:
        return ({"status" : "unchanged", "html_code" : html_txt})
    else:
        modified_html = find_text_in_between_tags(response['text'], start_tag="<!DOCTYPE html>", end_tag="</html>", inclusive=True)
        return ({"status" : "modified", "html_code" : response['text']})
````
